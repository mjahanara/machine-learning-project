```python
# importing lib

import pandas as pd
import numpy as np
from apyori import apriori
```

```python
super_market = pd.read_csv(r"E:\tmu . machine\EX\Dataset (1)\Dataset\supermarket.csv")
```

```python
super_market=super_market.replace('?', np.nan) 

super_market
```

```python
print(super_market.isnull().sum())
super_market.isnull().sum().sum()
```

```python
super_market = super_market.drop(['total'],axis=1)
super_market
```

```python
departments = list(super_market.columns)
sales = super_market.to_numpy().astype(str)
sales
```

```python
#Empty lists to be filled with the departments from which a customer has purchased at least one product
items = [ [] for _ in range(super_market.shape[0]) ]

#We iterate over every element in the sales matrix and, if a department belongs to a transaction, we append it
for i in range(super_market.shape[0]):
    for j in range(super_market.shape[1]):
        if sales[i,j] == "t":
            items[i].append(departments[j])
```

```python
min_supports = [0.1, 0.3, 0.5]
min_confidences = [0.9, 0.7, 0.5]

results = []

for i in min_supports:
    for j in min_confidences:
        result = list(apriori(items, min_support=i, min_confidence=j))
        results.append((i, j, result))
```

```python
for result in results:
    min_support, min_confidence, rules = result
    print(f'Rules with minimum support of {min_support} and minimum confidence of {min_confidence}:\n')
    for i in range(len(rules)):
        for j in range(len(rules[i][2])):
            print(list(rules[i][2][j][0]), '---->', list(rules[i][2][j][1]),
                  '\tsupp=' + str(round(rules[i][1], 2)) +
                  '\tconf=' + str(round(rules[i][2][j][2], 2)))
        print('\n')
```
